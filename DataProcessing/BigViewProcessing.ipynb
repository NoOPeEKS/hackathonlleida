```python
import pandas as pd
import json
from itertools import chain
from collections import ChainMap
```

Obrim un json de resposta de l'api d'inteligence

```python
with open('response.json') as f:
    data_dict = json.load(f)
```

Extraiem la part que ens interessa mostrar, els comparables

```python
data_list = data_dict['response']['comparables']
```

Preparem un diccionari amb llistes de cada atribut per facilitar despres la neteja i preparar el dataframe

```python
grouped_dict = {}
[grouped_dict.setdefault(key, []).append(value) for d in data_list for key, value in d.items()]
print(grouped_dict)
```

Ara agrupem els elements per posici√≥ i eliminem els elements que no ens interessen

```python
ordered_dict = {}
for key, value in grouped_dict.items():
    for i, inner_dict in enumerate(value):
        if key == 'media':
            continue
        ordered_dict.setdefault(i, []).append(inner_dict)
```

Ara desfem els diccionaris petits i ho posem tot en un diccionari gran

```python
final_dict = {}
for key, value in ordered_dict.items():
    final_dict[key] = dict(ChainMap(*value))
```

Finalment creem el dataframe i el trasposem per deixar les columnes con les volem

```python
df = pd.DataFrame.from_dict(final_dict).T
```
