```python
import pandas as pd
```

```python
def create_pandas(csv_file):
    return pd.read_csv(csv_file)
```

```python
dataset = create_pandas('./pisos.csv')
```

```python
dataset.set_index('id', inplace=True)
```

# DATASET DESCRIPTION

We have tons of data about houses located in Spain. The information we have is: 
+ **summary**: key words for searching in internet
+ **description**: brief description about the place
+ **location**
+ **photo**: url of the photo
+ **recomendado**: categorical variable. Set to "Recomendado" if the home is recomendated
+ **price**
+ **size**
+ **rooms**: number of rooms the house has
+ **price/m2**: relation between the price and the m2
+ **bathroooms**: number of bathrooms
+ **numphotos**
+ **type**
+ **region**

```python
wrong_items = list()

for i, j, z in zip(dataset['recomendado'], dataset['price'], dataset.index):
    if str(i)[-1] == '€':
        wrong_items.append(z)

dataset = dataset[~dataset.index.isin(wrong_items)]
```

```python
dataset['location'].isnull().sum()
```

```python
dataset = dataset.dropna(subset=['location'])
dataset['location'].isnull().sum()
```

```python
dataset['location'].nunique()
```

```python
dataset['region'].nunique()
```

```python
dataset['type'].unique()
```

```python
dataset.loc[dataset['type'] == 'pisos', 'type'] = 'Apartment'
dataset.loc[dataset['type'] == 'casas', 'type'] = 'House'
dataset.loc[dataset['type'] == 'piso', 'type'] = 'Apartment'
dataset.loc[dataset['type'] == 'aticos', 'type'] = 'Penthouses'
dataset.loc[dataset['type'] == 'duplex', 'type'] = 'Duplexes'
dataset.loc[dataset['type'] == 'estudios', 'type'] = 'Studios'
dataset.loc[dataset['type'] == 'fincas_rusticas', 'type'] = 'Rural Estates'
dataset.loc[dataset['type'] == 'lofts', 'type'] = 'Loft'

dataset
```

```python
urls = dataset['photo'].values

with open('urls.csv', 'w') as file:
    for url in urls:
        file.write(str(url))
        file.write('\n')
```
